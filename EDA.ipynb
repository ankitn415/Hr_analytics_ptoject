## Data Preview

This dataset contains HR-related information for employees, with the goal of analyzing **employee attrition** (whether employees left the company or not). The column `Attrition_Flag` is 1 if the employee left, and 0 otherwise.


```python
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(x="Attrition_Flag", data=df, palette="Set2")
plt.title("Attrition Class Distribution")
plt.xticks([0, 1], ["No", "Yes"])
plt.show()
```

## Class Balance

The dataset is **imbalanced**, with significantly more employees who stayed (`Attrition_Flag = 0`) than those who left (`Attrition_Flag = 1`). This should be taken into account during model training.


```python
correlation = df.corr(numeric_only=True)["Attrition_Flag"].sort_values(ascending=False)

# Visualizing top correlations
correlation.drop("Attrition_Flag").plot(kind="barh", color="teal")
plt.title("Correlation of Features with Attrition")
plt.xlabel("Correlation Coefficient")
plt.show()
```

## Correlation Insights

The most negatively correlated features (i.e., those that **reduce** the chances of attrition) include:

- **Job Level**
- **Total Working Years**
- **Years at Company**
- **Monthly Income**
- **Age**

These suggest that **older, well-paid, senior employees with more experience are less likely to leave**.


```python
categorical_cols = ['BusinessTravel', 'Department', 'EducationField', 
                    'Gender', 'JobRole', 'MaritalStatus', 'OverTime']

for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue="Attrition", palette="Set2")
    plt.title(f"{col} vs Attrition")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Categorical Analysis Insights

- **OverTime**: Employees working overtime are much more likely to leave.
- **JobRole**: Sales Representatives and Laboratory Technicians show higher attrition.
- **MaritalStatus**: Single employees are more likely to leave.
- **BusinessTravel**: Frequent travelers show high attrition.


```python
numerical_cols = ['Age', 'DistanceFromHome', 'MonthlyIncome', 
                  'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany']

for col in numerical_cols:
    plt.figure(figsize=(10, 4))
    sns.kdeplot(data=df, x=col, hue="Attrition", fill=True)
    plt.title(f"Distribution of {col} by Attrition")
    plt.show()
```

##  Numeric Feature Insights

- **MonthlyIncome**: Employees with lower salaries tend to leave more.
- **DistanceFromHome**: People living farther have slightly higher attrition.
- **YearsAtCompany**: Newer employees are more likely to quit.


```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```
